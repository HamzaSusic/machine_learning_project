```python
from scipy import stats
import numpy as np
```

```python
#Checking for the outliers by using z-score instead of visualization
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols].apply(stats.zscore)
```

```python
z_scores = df[numeric_cols].apply(stats.zscore)
threshold = 5 #Not sure for threshold
mask = (z_scores < threshold).all(axis=1) & (z_scores > -threshold).all(axis=1)
```

```python
df_wo_outliers = df[mask]
df_wo_outliers
```

```python
#Creating a copy of dataframe and creating Simple Moving Average over the 10 days span
#10 days is default window span since majority of stock markets are working 5 days a week
#Thus this represents market and exchanges on 2 working weeks
df_wo_outliers_copy = df_wo_outliers.copy()
df_wo_outliers_copy['SMA_10'] = df_wo_outliers['Close'].rolling(window=10).mean()
df_wo_outliers_copy
```

```python
def rsi_calculation (data, window):
    delta = data['Close'].diff(1)
    gain = delta.where(delta>0, 0)
    loss = -delta.where(delta <0, 0)

    mean_gain = gain.rolling(window=window, min_periods=window).mean()
    mean_loss = loss.rolling(window=window, min_periods=window).mean()
    
    rs = mean_gain / mean_loss
    rsi = 100 - (100/ (1+rs))
    return rsi
df_wo_outliers_copy['RSI'] = rsi_calculation(df_wo_outliers_copy, 14)
df_wo_outliers_copy
```

```python
df_wo_outliers_copy['RSI'].describe()
```

```python
def bollinger_bands_calc(data, window, num_std_dev):
    rolling_mean = data['Close'].rolling(window=window).mean()
    rolling_std = data['Close'].rolling(window=window).mean()
    upper_band = rolling_mean + (rolling_std * num_std_dev)
    lower_band = rolling_mean - (rolling_std * num_std_dev)
    return rolling_mean, upper_band, lower_band

window= 20 #Default one
num_std_dev = 2
df_wo_outliers_copy['Mid Band'], df_wo_outliers_copy['Upper Band'], df['Lower_Band'] = bollinger_bands_calc(df_wo_outliers_copy, window, num_std_dev)
df_wo_outliers_copy
```

```python
def macd_calc (data, short_window, long_window, signal_window):
    short_ema = data['Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = data['Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal

short_window = 12
long_window = 26
signal_window = 9

df_wo_outliers_copy['MACD'], df_wo_outliers_copy['Signal Line'] = macd_calc(df_wo_outliers_copy, short_window, long_window, signal_window)
df_wo_outliers_copy
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

df_wo_outliers_copy = df_wo_outliers_copy.fillna(0)  # Replace NaNs with 0

# Select relevant features for multicollinearity analysis
features = ['Close', 'Volume', 'SMA_10', 'EMA_10', 'Volatility', 'RSI', 'Mid Band', 'Upper Band', 'MACD', 'Signal Line']
X = df_wo_outliers_copy[features]
 
# Calculate VIF
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
 
print(vif_data)
```

```python
#VIF confirmed correlation between Mid Band and Upper Band, as well as correlation between SMA_10 and EMA, and irrelevant feature like Volume
```
