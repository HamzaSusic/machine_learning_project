```python
from scipy import stats
import numpy as np
```

```python
#Checking for the outliers by using z-score instead of visualization
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols].apply(stats.zscore)
```

```python
z_scores = df[numeric_cols].apply(stats.zscore)
threshold = 5 #Not sure for threshold
mask = (z_scores < threshold).all(axis=1) & (z_scores > -threshold).all(axis=1)
```

```python
df_wo_outliers = df[mask]
df_wo_outliers
```

```python
#Creating a copy of dataframe and creating Simple Moving Average over the 10 days span
#10 days is default window span since majority of stock markets are working 5 days a week
#Thus this represents market and exchanges on 2 working weeks
df_wo_outliers_copy = df_wo_outliers.copy()
df_wo_outliers_copy['SMA_10'] = df_wo_outliers['Close'].rolling(window=10).mean()
df_wo_outliers_copy
```
