```python
#Importing pandas library for reading and using different functions with dataframe
import pandas as pd
```

```python
#Reading the dataset and displaying it
df = pd.read_csv('Stocks.csv')
df
```

```python
#Getting info about the data in dataset
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print(df['Date'].dtype)
```

```python
df.info()
```

```python
#Getting mean, mode, median, variation..
df.describe()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Since the Capital Gains column is almost fully empty, we can drop it
df=df.drop(columns=['Capital Gains'])
df
```

```python
df.info()
```

```python
df['Rounded_Close'] = df['Close'].round(2)
df['Rounded_Open'] = df['Open'].round(2)
df['Rounded_High'] = df['High'].round(2)
df['Rounded_Low'] = df['Low'].round(2)
df
```

```python
from scipy import stats
import numpy as np
```

```python
#Checking for the outliers by using z-score instead of visualization
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols].apply(stats.zscore)
```

```python
z_scores = df[numeric_cols].apply(stats.zscore)
threshold = 5 #Not sure for threshold
mask = (z_scores < threshold).all(axis=1) & (z_scores > -threshold).all(axis=1)
```

```python
df_wo_outliers = df[mask]
df_wo_outliers
```

```python
#Creating a copy of dataframe and creating Simple Moving Average over the 10 days span
#10 days is default window span since majority of stock markets are working 5 days a week
#Thus this represents market and exchanges on 2 working weeks
df_wo_outliers_copy = df_wo_outliers.copy()
df_wo_outliers_copy['SMA_10'] = df_wo_outliers['Close'].rolling(window=10).mean()
df_wo_outliers_copy
```

```python
df_wo_outliers_copy['SMA_10']
```

```python
#Exponential Moving Average over 10 days
#Recent price changes
df_wo_outliers_copy['EMA_10'] = df_wo_outliers_copy['Close'].ewm(span=10, adjust=False).mean()
df_wo_outliers_copy
```

```python
import matplotlib.pyplot as plt
```

```python
df_wo_outliers_copy['Date'] = pd.to_datetime(df_wo_outliers_copy['Date']).dt.tz_localize(None)
```

```python
plt.figure (figsize=(14,7))

plt.plot(df_wo_outliers_copy['Date'], df_wo_outliers_copy['Close'], label='Closing price', color='blue')
plt.plot(df_wo_outliers_copy['Date'], df_wo_outliers_copy['SMA_10'], label='SMA over 10 days', color='red')
#plt.plot(df_wo_outliers_copy['Date'], df_wo_outliers_copy['EMA_10'], label='EMA over 10 days', color='green')

plt.title('Close price with SMA', fontsize=15)
plt.xlabel('Date', fontsize=11)
plt.ylabel('Price', fontsize=11)

plt.legend(fontsize=12)
plt.grid(True, alpha=0.2)
plt.show()
```

```python
#Adding new feature 'Volatility'
#Stock price fluctuation over certain time (10 days)
df_wo_outliers_copy['Volatility'] = df_wo_outliers_copy['Close'].rolling(window=10).std()
df_wo_outliers_copy
```

```python
import seaborn as sns

#Since we have non-numerical data, we chose to use only numerical one for easier representation
df_num = df_wo_outliers_copy.select_dtypes(include=['number'])

#Once the numerical values are only considered, we did correlation matrix for feature selection
corr_matrix=df_num.corr()

#Visualization of correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix with only numerical values")
plt.show()
```

```python
#Open, High, Low, Close highly correlated with one another (expected)
#Moving averages we added have medium to strong correlation with 'Close' (Good feature)
#Dividends and Stock Splits - weak or almost no correlation with 'Close' (Candidates for dropping)
#Volatility - correlated with moving averages

#Due to high correlation for 4 main features, we could reduce redundancy and use only one or two variables
#Or even create new feature like average of them or range
#We plan to do additional techniques for feature selection
```
