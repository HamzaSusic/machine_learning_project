```python
import pandas as pd
import numpy as np

# Sample data for 2020
data = {
    'Date': pd.date_range(start='1/1/2020', periods=366, freq='D'),
    'Close': pd.Series(range(366)) + 100,
    'ema_10': pd.Series(range(366)) + 50,
    'volatility': pd.Series(range(366)) * 0.1,
    'RSI': pd.Series(range(366)) * 0.05,
    'Mid Band': pd.Series(range(366)) * 0.2,
    'MACD': pd.Series(range(366)) * 0.3,
    'signal line': pd.Series(range(366)) * 0.4,
    'close_log': pd.Series(range(366)).apply(lambda x: x if x == 0 else np.log(x))
}

df_2020 = pd.DataFrame(data)

# Feature Engineering

# 1. Lag Features
df_2020['Close_lag_1'] = df_2020['Close'].shift(1)
df_2020['Close_lag_2'] = df_2020['Close'].shift(2)
df_2020['Close_lag_3'] = df_2020['Close'].shift(3)

# 2. Rolling Statistics
df_2020['Close_roll_mean_5'] = df_2020['Close'].rolling(window=5).mean()
df_2020['Close_roll_std_5'] = df_2020['Close'].rolling(window=5).std()
df_2020['ema_10_roll_mean_5'] = df_2020['ema_10'].rolling(window=5).mean()
df_2020['ema_10_roll_std_5'] = df_2020['ema_10'].rolling(window=5).std()
df_2020['volatility_roll_mean_5'] = df_2020['volatility'].rolling(window=5).mean()
df_2020['volatility_roll_std_5'] = df_2020['volatility'].rolling(window=5).std()

# 3. Interaction Features
df_2020['ema_10_volatility'] = df_2020['ema_10'] * df_2020['volatility']
df_2020['RSI_Mid_Band'] = df_2020['RSI'] * df_2020['Mid Band']
df_2020['MACD_signal_line'] = df_2020['MACD'] * df_2020['signal line']

print(df_2020.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Create interaction features (if not already in the dataset)
# For demonstration, assuming the interaction features are already calculated
features = ['ema_10_volatility', 'RS1_Mid_Band', 'MACD_signal_line']  # Replace with actual column names
target_price = 'Close'

# Classification target: 1 if price increases, 0 otherwise
df['Price_Direction'] = np.where(df[target_price].shift(-1) > df[target_price], 1, 0)

# Regression target: Predict exact price
target_classification = 'Price_Direction'
target_regression = target_price

# Train-test split (80-20)
# Selecting the columns corresponding to the features
X = df[features]  # Ensure 'features' contains column names as strings

# Target variable for classification and regression
y_classification = df[target_classification]  # Ensure this is a valid column
y_regression = df[target_regression]  # Ensure this is a valid column

X_train, X_test, y_train_class, y_test_class = train_test_split(X, y_classification, test_size=0.2, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train_reg = scaler.fit_transform(X_train_reg)
X_test_reg = scaler.transform(X_test_reg)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train_class)

# Predictions
y_pred_class = clf.predict(X_test)

# Evaluate model
print("Classification Report:")
print(classification_report(y_test_class, y_pred_class))
print(f"Accuracy: {accuracy_score(y_test_class, y_pred_class) * 100:.2f}%")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data = df

# Convert date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Filter data for 2021 and 2022 for training
train_data = data[(data['Date'] >= '2021-01-01') & (data['Date'] <= '2022-12-31')]

# Filter data for 2023 for testing
test_data = data[(data['Date'] >= '2023-01-01') & (data['Date'] <= '2023-12-31')]

# Select numerical features only
numerical_features = ['Close', 'EMA_10', 'Volatility', 'RSI', 'Mid Band', 'MACD', 'Signal Line', 
                      'Close_log', 'Close_lag_1', 'Close_lag_2', 'Close_lag_3', 
                      'Close_roll_mean_5', 'Close_roll_std_5', 'ema_10_roll_mean_5', 
                      'ema_10_roll_std_5', 'volatility_roll_mean_5', 'volatility_roll_std_5', 
                      'ema_10_volatility', 'RS1_Mid_Band', 'MACD_signal_line']

# Prepare features and target for training
train_features = train_data[numerical_features].dropna()
train_target = train_data['Close'].shift(-1).dropna()

# Align features and target for training
train_features = train_features.iloc[:-1]
train_target = train_target.iloc[1:]

# Ensure the same length of features and target
min_length = min(len(train_features), len(train_target))
train_features = train_features.iloc[:min_length]
train_target = train_target.iloc[:min_length]

# Prepare features and target for testing
test_features = test_data[numerical_features].dropna()
test_target = test_data['Close'].shift(-1).dropna()

# Align features and target for testing
test_features = test_features.iloc[:-1]
test_target = test_target.iloc[1:]

# Ensure the same length of features and target
min_length_test = min(len(test_features), len(test_target))
test_features = test_features.iloc[:min_length_test]
test_target = test_target.iloc[:min_length_test]

# Split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(train_features, train_target, test_size=0.2, random_state=42)

# Train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on validation set
val_predictions = model.predict(X_val)

# Evaluate on validation set
mse = mean_squared_error(y_val, val_predictions)
print(f'Mean Squared Error on validation set: {mse}')

# Predict on test set (2023 data)
test_predictions = model.predict(test_features)

# Calculate mean close price for 2023
mean_close_2023 = test_data['Close'].mean()

print(f'Mean close price for 2023: {mean_close_2023}')
print(f'Predicted next day price: {test_predictions[-1]}')
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Convert date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Filter data for 2021 and 2022 for training
train_data = data[(data['Date'] >= '2021-01-01') & (data['Date'] <= '2022-12-31')]

# Filter data for 2023 for testing
test_data = data[(data['Date'] >= '2023-01-01') & (data['Date'] <= '2023-12-31')]

# Select numerical features only
numerical_features = ['Close', 'EMA_10', 'Volatility', 'RSI', 'Mid Band', 'MACD', 'Signal Line', 
                      'Close_log', 'Close_lag_1', 'Close_lag_2', 'Close_lag_3', 
                      'Close_roll_mean_5', 'Close_roll_std_5', 'ema_10_roll_mean_5', 
                      'ema_10_roll_std_5', 'volatility_roll_mean_5', 'volatility_roll_std_5', 
                      'ema_10_volatility', 'RS1_Mid_Band', 'MACD_signal_line']

# Prepare features and target for training
train_features = train_data[numerical_features].dropna()
train_target = train_data['Close'].shift(-1).dropna()

# Align features and target for training
train_features = train_features.iloc[:-1]
train_target = train_target.iloc[1:]

# Ensure the same length of features and target
min_length = min(len(train_features), len(train_target))
train_features = train_features.iloc[:min_length]
train_target = train_target.iloc[:min_length]

# Prepare features and target for testing
test_features = test_data[numerical_features].dropna()
test_target = test_data['Close'].shift(-1).dropna()

# Align features and target for testing
test_features = test_features.iloc[:-1]
test_target = test_target.iloc[1:]

# Ensure the same length of features and target
min_length_test = min(len(test_features), len(test_target))
test_features = test_features.iloc[:min_length_test]
test_target = test_target.iloc[:min_length_test]

# Scale the data
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

scaled_train_features = scaler_X.fit_transform(train_features)
scaled_train_target = scaler_y.fit_transform(train_target.values.reshape(-1, 1))

scaled_test_features = scaler_X.transform(test_features)
scaled_test_target = scaler_y.transform(test_target.values.reshape(-1, 1))

# Reshape the data to fit LSTM input requirements [samples, time steps, features]
X_train = np.reshape(scaled_train_features, (scaled_train_features.shape[0], 1, scaled_train_features.shape[1]))
y_train = scaled_train_target

X_test = np.reshape(scaled_test_features, (scaled_test_features.shape[0], 1, scaled_test_features.shape[1]))
y_test = scaled_test_target

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Predict on test set (2023 data)
predictions = model.predict(X_test)

# Inverse transform the predictions to get actual values
predicted_prices = scaler_y.inverse_transform(predictions)

# Calculate mean close price for 2023
mean_close_2023 = test_data['Close'].mean()

print(f'Mean close price for 2023: {mean_close_2023}')
print(f'Predicted next day price: {predicted_prices[-1][0]}')
```
